```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

pd.options.display.float_format = "{:,.2f}".format
```

```python
# Finding top 10 countries with the highest population for the year 2022

population.sort_values(by = '2022 Population', ascending = False).head(10)
```

```python
# Finding the top 10 countries based on world population percentage

population.sort_values(by = 'World Population Percentage', ascending = False).head(10)
```

```python
corr_matrix = population.corr(numeric_only = True)
corr_matrix
```

```python
deepblue = sns.color_palette("light:#5A9", as_cmap=True)
sns.heatmap(data = corr_matrix, annot = True, cmap = deepblue)
plt.rcParams['figure.figsize'] = (20, 7)
plt.show(block = True)
```

```python
# Are there certain continents that have grown faster than others and in which ways?

population2 = population.groupby('Continent').mean(numeric_only = True).sort_values(by = '2022 Population', ascending = False)
```

```python
# This is just visualizing all the columns. Note that this is incorrect, just wanted to test this plot() function
population3 = population2.transpose()
population3.plot()
```

```python
population2 = population.groupby('Continent')[population.columns[12:4:-1]].mean(numeric_only = True).sort_values(by = '2022 Population', ascending = False)
population2
```

```python
population2.transpose().plot()
```

```python
sns.set_theme(style = 'darkgrid')
sns.lineplot(data = population2.transpose(), dashes = False).set(title = 'Population by Continent', xlabel = 'Years', ylabel = 'Population')
```

```python
# Creating a boxplot for this dataset

population3.boxplot(figsize = (20, 10)).set(title = 'Boxplot for Population by Year')
```
